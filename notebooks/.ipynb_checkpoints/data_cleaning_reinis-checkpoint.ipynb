```python
def clean_data(data):
    '''
    Returns cleaned dataframe
    '''
    # Dropping almost empty columns and guest_id
    data = data.drop(columns = ['entry_date_hour', 'exit_date_hour', 'guest_id'])
    
    # Leaving only popular products [H10, F397, F398, F44, F98, F60, F63]
    product_names = ['H10', 'F397', 'F398', 'F44', 'F98', 'F60', 'F63']
    data = data[data['product'].isin(product_names)]
    
    # Leaving 4 chosen pockets 
    pockets_names = ['0000000095', '0000100368', '0000000052', '0000100435']
    data = data[data['pocket'].isin(pockets_names)]
    
    # Dropping columns whith option electrique
    data = data[data['option']!='electrique']
    
    #Leaving only the 5 most popular promo codes
    promo_codes = ['VOYAGEURSEMAINE', 'OPFRAISDERESA0PREM', 'ETE2019', 'ETE2021AUTO',
       'VOYAGEURWE']
    #data = data[data['promo_code'].isin(promo_codes)]
    
    
    return data
```
